```python
#import different databases
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats.stats import pearsonr #We will need the correlation function from the scientific package scipy
```

```python
#I make a dataframe of the CSV file and show the first 5 results
df = pd.read_csv('funda-amsterdam.csv')
df.head()
```

```python
#I make a subset to leave out the prices under 30000. These are the outliers of the dataset
df_subset = df[df['price'] > 30000]
df_subset = df_subset[['price', 'area', 'rooms']]
```

```python
#I make a scatterplot of the area and price
sns.scatterplot(data=df_subset, x="price", y="area")
```

As you can see in the scatterplot above, there is a correlation between area and price. The data is going up linear.

```python
#I make a pairplot with three different quantitative variables
sns.pairplot(df_subset)
```

In the pairplot are multiple scatterplots. All data is going up linear.

```python
#I make a scatterplot two variables.
sns.scatterplot(data=df_subset, x="price", y="rooms")
```

The correlation is weak but it is still linear.

```python
#I calculate the pearsons r
pearsonr(df_subset['price'],df_subset['rooms'])
```

The output gives us two numbers: r which is 0.54, and p which appears to be 0, but which is actually just really small. We can conclude the following:

Subscribers and views are moderate (r = 0.54)
The correlation is highly significant (p < 0.001)
