```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv('surveydata.csv')
df1.head()
```

```python
df2 = pd.read_csv('stepsdata.csv', sep=';')
df2.head()
```

```python
df = pd.merge(df1, df2, on= 'id') #'id' is the common identifier
df.plot(y="waist")
```

```python
df["waist"].value_counts()
```

```python
df['waist'] = df['waist'].apply(pd.to_numeric, errors='coerce')
#To remove the values above 45 we need to create a function that we then apply to the entire column
def between_45and130(x):
    if(x < 45.0 or x > 130): 
        return float('NaN')
    else: 
        return x
df['waist'] = df['waist'].apply(between_45and130)
```

```python
df.plot(y="waist", x = "id", kind='scatter')
```
